# Preprocessing (Encoding & Scaling)

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load processed data
df = pd.read_csv("../data/feature_engineering_Processed.csv")
df.head()
```

## Encode Categorical Variables

```python

# Encode 'sex', 'source', 'browser', 'country'
label_encoders = {}
for col in ['sex', 'source', 'browser', 'country']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

df[['sex', 'source', 'browser', 'country']].head()
    
```

## Scale Numerical Features

```python

scaler = StandardScaler()

# Select columns to scale
num_cols = ['purchase_value', 'age', 'time_since_signup']

df[num_cols] = scaler.fit_transform(df[num_cols])
df[num_cols].head()
    
```

## Save Preprocessed Data

```python

df.to_csv("../data/final_fraud_data_Processed.csv", index=False)    
```
